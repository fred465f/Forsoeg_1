```python
# Load experiment data
df_glas_til_luft = pd.read_csv('Samlet_glas_til_luft_df.csv')
df_luft_til_glas = pd.read_csv('Samlet_luft_til_glas_df.csv')
```

```python
# ------------------------------------------------------------
# Glass to air experiment data transformation
# ------------------------------------------------------------

# Define function to let Angle_big --> Angle_2 (theta_2)
def convert_gta(x):
    return x[0] - (np.pi - x[1])

# Define function to transform angles --> sine(angles * 2pi/360)
def sine_gta(x): 
    return np.sin(x * 2*np.pi/360)

# Define function to propagate error for Angle_2 --> sine(Angle_2 * 2pi/360)
def error_prop_gta(x): 
    return (2*np.pi/360 * np.cos(x[1] * 2*np.pi/360))**2 * (x[2])**2

# Apply functions to data
df_glas_til_luft['Angle_2'] = df_glas_til_luft.apply(convert_gta, axis = 1)
df_glas_til_luft['Angle_2_sine'] = df_glas_til_luft['Angle_2'].apply(sine_gta)
df_glas_til_luft['Angle_small_sine'] = df_glas_til_luft['Angle_small'].apply(sine_gta)
df_glas_til_luft['Angle_alpha_sine_prop'] = df_glas_til_luft.apply(error_prop_gta, axis = 1)
```

```python
# ------------------------------------------------------------
# Air to glass experiment data transformation
# ------------------------------------------------------------

# Define function to let Angle_big --> Angle_2 (theta_2)
def convert_atg(x):
    return x[0] - (np.pi - x[1])

# Define function to transform angles --> sine(angles * 2pi/360)
def sine_atg(x): 
    return np.sin(x * 2*np.pi/360)

# Define function to propagate error for Angle_2 --> sine(Angle_2 * 2pi/360)
def error_prop_atg(x): 
    return (2*np.pi/360 * np.cos(x[1] * 2*np.pi/360))**2 * (x[2])**2

# Apply functions to data
df_luft_til_glas['Angle_2'] = df_glas_til_luft.apply(convert_atg, axis = 1)
df_luft_til_glas['Angle_2_sine'] = df_glas_til_luft['Angle_2'].apply(sine_atg)
df_luft_til_glas['Angle_small_sine'] = df_glas_til_luft['Angle_small'].apply(sine_atg)
df_luft_til_glas['Angle_alpha_sine_prop'] = df_glas_til_luft.apply(error_prop_atg, axis = 1)
```

```python
# Add experiments to virtual lab
lab.add_experiment(name = 'glas_til_luft', data = df_glas_til_luft)
lab.add_experiment(name = 'luft_til_glas', data = df_luft_til_glas)
```

```python
# Define model for 'glas_til_luft' and 'luft_til_glas'
n_air = 1
def theta_1_gta(theta_2, n_glass):
    return n_air/n_glass * theta_2
def theta_1_atg(theta_2, n_glass):
    return n_glass/n_air * theta_2

# Define variables
columns = [['Angle_2_sine', 'Angle_small_sine']]
funcs_gta = [theta_1_gta]
funcs_atg = [theta_1_atg]
errors_gta = [df_glas_til_luft['Angle_alpha_sine_prop']]
errors_atg = [df_luft_til_glas['Angle_alpha_sine_prop']]

# Fit for refractive index of optic glass
popts_gta, perrs_gta = lab.fit(name = 'glas_til_luft', columns = columns, 
        funcs = funcs_gta, guesses=1.5, sigma=errors_gta, 
        absolute_sigma=True, residuals=True, 
        save=False, alpha=1)
popts_atg, perrs_atg = lab.fit(name = 'luft_til_glas', columns = columns, 
        funcs = funcs_atg, guesses=1.5, sigma=errors_atg, 
        absolute_sigma=True, residuals=True, 
        save=False, alpha=1)
```

```python
# Calculate refrative index of glass directly
df_glas_til_luft['n_glas'] = df_glas_til_luft.apply(lambda x: x[12]/x[13], axis = 1)
df_luft_til_glas['n_glas'] = df_luft_til_glas.apply(lambda x: x[13]/x[12], axis = 1)
n_glas_gta = df_glas_til_luft['n_glas'].mean()
n_glas_atg = df_luft_til_glas['n_glas'].mean()
print(n_glas_gta, n_glas_atg)
```
