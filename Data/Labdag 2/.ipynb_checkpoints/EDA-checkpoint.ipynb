```python
df = pd.read_csv('Samlet_df.csv')
lab.add_experiment(data = df, name = 'Coff_data')
df
```

```python
lab.analysis(name = 'Coff_data', columns=[['Angle', 'Coff_Rp']], analysis_type="Scatter", save=False)
```

```python
def model(x, theta_2):
    return np.tan(x - theta_2)**2/np.tan(x + theta_2)**2

lab.fit(name = 'Coff_data', columns = [['Angle', 'Coff_Rp']], funcs = [model], guesses = 0.2, sigma=[df['Coff_Rp_err']], 
        absolute_sigma=True, residuals=True, save=True)
```

```python
n_glas = 2

def calc_theta_2(x):
    return np.arcsin((x[0] * 2*np.pi)/(360 * n_glas)) * (360/(2*np.pi))

df['Angle_2'] = df.apply(calc_theta_2, axis = 1)
```

```python
df
```

```python
def calc_Rp(x):
    return np.tan(x[0] - x[-1])**2/np.tan(x[0] + x[-1])**2

Calc_Rp = df.apply(calc_Rp, axis = 1)
Calc_Rp
```
